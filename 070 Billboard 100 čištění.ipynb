```python
import os
import pandas as pd
```

```python
df = pd.read_csv(os.path.join("data_b100","charts.csv"))
```

```python
df
```

```python
df['rok'] = pd.to_numeric(df['date'].apply(lambda x: x[0:4]))
```

```python
df['rok']
```

```python
df['datum'] = df['date'].apply(lambda x: x[5:])
```

```python
df = df.sort_values(by=['rok','datum'])
```

```python
df
```

```python
df
```

```python
data = df[['rok','datum']].drop_duplicates()
```

```python
data
```

```python
data['tyden'] = data.groupby('rok').cumcount() + 1
```

```python
df = df.merge(data, on=['rok','datum'])
```

```python
df
```

```python
df = df[df['rok'] > 1960]
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
spravna_jmena = {
    "Beyonce": "Beyoncé",
    "The Chainsmokers": "Chainsmokers",
    "The Black Eyed Peas": "Black Eyed Peas"
}
```

```python
for key, val in spravna_jmena.items():
    df['artist'] = df['artist'].apply(lambda x: x.replace(key,val))
```

```python
def kdojeprvni(x):
    delice = ["&","/","(feat.","feat.","Feat."," a "," x "," X "," with "," With "," featuring "," Featuring",", "," + "]
    for d in delice:
        if d in x:
            x = x.split(d)[0]
    return x.strip()
```

```python
df['interpret_hlavni'] = df['artist'].apply(lambda x: kdojeprvni(x))
```

```python
df
```

```python
df.to_csv(os.path.join("data_b100","b100.csv"), index=False)
```

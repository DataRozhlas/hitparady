```python
import os
import json
import pandas as pd
```

```python
def load_json_file(file):
    with open(file) as f:
        data = json.load(f)
    return data['data']
```

```python
directory = "radiozurnal"
file_list = os.listdir(directory)
```

```python
df = pd.DataFrame()
frames = []
for file in file_list:
    try:
        if file.endswith(".json"):
            filepath = os.path.join(directory, file)
            dataframe = pd.DataFrame(load_json_file(filepath))
            frames.append(dataframe)
    except:
        print(file)
```

```python
df = pd.concat(frames)
```

```python
df = df.drop(columns=['files','itemcode']).reset_index(drop=True)
```

```python
df
```

```python
slovodelice = [" ",".","/","'"]
for s in slovodelice:
    df['interpret'] = df['interpret'].apply(lambda x: s.join(word.title() for word in str(x).split(s)))
    df['track'] = df['track'].apply(lambda x: s.join(word.title() for word in str(x).split(s)))
```

```python
df
```

```python
df.to_json(os.path.join("data","radiozurnal.json"), orient="records")
```

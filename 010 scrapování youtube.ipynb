```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
youtube = pd.read_csv(os.path.join("data","youtube_linky.csv"))
```

```python
youtube = youtube.dropna()
```

```python
youtube
```

```python
def delka_videa(url):
    # lol to je hnusná funkce, ale spěchal jsem
    if "watch" in url:
        
        url = url.split("&")[0]
        
        if url[0:3] == "www":
            url = "https://" + url
        if url[0:5] == "http:":
            url = url.replace("http:","https:")
        
        try:
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            soup = str(soup)
            if "chapterRenderer" in soup:
                return None
            if "approxDurationMs" in soup:
                return int(soup.split("approxDurationMs")[1].split("\"")[2]) / 1000
            else:
                print(url)
                return None
        except:
            print(url)
            return None
    else:
        return None
```

```python
youtube['delka'] = youtube['youtube'].apply(lambda x: delka_videa(x))
```

```python
import os
import pandas as pd
```

```python
cz = pd.read_csv(os.path.join("data","CZ_RADIO_TOP_100_tydenni.csv"))
sk = pd.read_csv(os.path.join("data","SK_RADIO_TOP_100_tydenni.csv"))
```

```python
cz = cz[cz['rok'] > 2017]
sk = sk[sk['rok'] > 2017]
```

```python
pomer = pd.Series(cz.groupby('interpret_hlavni').size() / sk.groupby('interpret_hlavni').size(), name='pomer')
cz_count = pd.Series(cz.groupby('interpret_hlavni').size(), name='cz_count')
sk_count = pd.Series(sk.groupby('interpret_hlavni').size(), name='sk_count')
```

```python
srovnani = pd.DataFrame([pomer,cz_count,sk_count]).transpose().fillna(0)
```

## Československý žebříček

```python
srovnani['total'] = srovnani['cz_count'] + srovnani['sk_count']
```

```python
srovnani[(srovnani['pomer'] > 0.25) & (srovnani['pomer'] < 0.75)].sort_values(by='total', ascending=False).head(10)
```

```python
bratrske_interpretstvo = srovnani[(srovnani['pomer'] > 0.25) & (srovnani['pomer'] < 0.75)].sort_values(by='total', ascending=False).head(10).index.to_list()
```

## Kdo nejvíc bodoval v ČR, ale v SR dostal do top 100 max 5×

```python
srovnani[srovnani['sk_count'] <= 5].sort_values(by='cz_count', ascending=False).head(20)
```

```python
", ".join(srovnani[srovnani['sk_count'] == 0].sort_values(by='cz_count', ascending=False).head(5).index.to_list())
```

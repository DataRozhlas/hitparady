```python
import os
```

```python
import pandas as pd
```

```python
folder = "data_raw"
```

```python
def seznam_souboru(slozka):
    seznam = []
    for file in os.listdir(slozka):
        seznam.append(file)
    return(seznam)
```

```python
raw_data = seznam_souboru(folder)
```

```python
from IPython.display import clear_output
```

```python
df = pd.DataFrame()
count = 0
for r in raw_data:
    count = count + 1
    print(f"done: {count} / total: {len(raw_data)}")
    df = pd.concat([df, pd.read_json(os.path.join(folder,r))])
    clear_output(wait=True)
```

```python
import os
import pandas as pd
```

```python
df = pd.read_csv(os.path.join('wikidata','wikidata_raw.csv'))
```

## Statistiky

```python
sloupce = df.columns.to_list()
```

```python
statistiky = []
for s in sloupce:
    statka = {}
    statka['sloupec'] = s
    statka['vyplneno'] = len(df[~df[s].isnull()])
    statka['nevyplneno'] = len(df[df[s].isnull()])
    statka['ruznych_hodnot'] = df[s].nunique()
    statistiky.append(statka)
statistiky = pd.DataFrame(statistiky)
statistiky.sort_values(by='nevyplneno')
```

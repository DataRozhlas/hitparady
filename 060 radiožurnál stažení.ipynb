```python
folder = "radiozurnal"
```

```python
import os

if not os.path.exists(folder):
    os.mkdir(folder)
else:
    pass
```

```python
import requests
```

```python
from datetime import date, timedelta
```

```python
start_dt = date(2020, 12, 1)
end_dt = date(2023, 10, 22)
delta = timedelta(days=1)
dates = []
while start_dt <= end_dt:
    dates.append(str(start_dt.isoformat()))
    start_dt += delta
```

```python
dates
```

```python
for d in dates:
    url = f"""https://api.rozhlas.cz/data/v2/playlist/day/{d.replace("-", "/")}/radiozurnal.json"""
    r = requests.get(url)
    with open(os.path.join(folder, f"rz-{d}.json"), 'w+', encoding='utf-8') as f:
        print(url)
        f.write(r.text)
```
